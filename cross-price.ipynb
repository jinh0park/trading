```python
import requests
from pprint import pprint
import json
```

```python
res = requests.get("https://api.upbit.com/v1/market/all")
res = json.loads(res.text)
#for y in filter(lambda x:x['market'].split('-')[0] == 'KRW', res): print(y['market'], end=', ')
krw_markets = map(lambda z: z['market'], filter(lambda x:x['market'].split('-')[0] == 'KRW', res))
res = requests.get("https://api.upbit.com/v1/ticker", params={"markets":','.join(krw_markets)})
res = json.loads(res.text)
res
```

```python
for x in sorted(res, key=lambda r: -r['acc_trade_price_24h'])[:5]:
    print(x['market'], x['acc_trade_price_24h'])
```
